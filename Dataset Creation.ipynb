```python
import pandas as pd
import unidecode
```

```python
drivers = pd.read_csv('dataset/drivers.csv')
races = pd.read_csv('dataset/races.csv')
teams = pd.read_csv('dataset/constructors.csv')
race_results = pd.read_csv('dataset/results.csv')
circuit_info = pd.read_csv('dataset/circuits.csv')
circuit_info = circuit_info[['circuitId', 'lat', 'lng']]


driver_race_results = pd.merge(race_results, drivers, on="driverId")
driver_race_results['driver_name'] = driver_race_results[['forename', 'surname']].apply(lambda x: unidecode.unidecode(' '.join(x)), axis=1)
driver_race_results = pd.merge(driver_race_results, races, on='raceId')
driver_race_results = pd.merge(driver_race_results, teams, on='constructorId')
driver_race_results = pd.merge(driver_race_results, circuit_info, on='circuitId')
driver_race_results.drop(['number_x', 'positionText', 'position', 'time_x', 'milliseconds', 'fastestLap', 'rank', 'statusId','number_y','code','url_x', 'url_y', 'time_y', 'url', 'forename', 'surname'], axis=1, inplace=True)
driver_race_results.rename(columns={'lng':'lon','grid': 'start_position', 'positionOrder': 'finish_position','nationality_x':'driver_nationality', 'nationality_y':'team_nationality', 'year':'race_year', 'name_x':'circuit_name', 'name_y':'team_name'}, inplace=True)
driver_race_results.sort_values('resultId', inplace=True)
driver_race_results.to_csv("race_results_mega.csv", index=False)
```
